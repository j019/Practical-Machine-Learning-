# Heart Disease : 0 and 1 ['V58']

```python
df['V58'].unique()
```

```python
X = df.drop(columns=['V58'])
Y = df['V58']
X.shape, Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=7,stratify=Y)

X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
Y.value_counts(normalize=True)
```

```python
Y_train.value_counts(normalize=True)
```

```python
Y_test.value_counts(normalize=True)
```

## For logistic Regression And SVM(compulsory) we have to do Standardize the data

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
X_train_sc.shape, X_test_sc.shape
```

```python
X_train_sc.mean(), X_test_sc.mean()
```

```python
X_train_sc.std(), X_test_sc.std()
```

## Apply ML Models

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(class_weight='balanced', random_state=7)
lr.fit(X_train_sc, Y_train)
Y_pred = lr.predict(X_test_sc)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_pred))
```

```python
# Gaussian Naive Bayes
from sklearn.naive_bayes import GaussianNB, BernoulliNB
gnb = GaussianNB()
gnb.fit(X_train_sc, Y_train)
Y_pred = gnb.predict(X_test_sc)
print(classification_report(Y_test, Y_pred))
```

```python
# Bernoulli Naive Bayes
bnb = BernoulliNB()
bnb.fit(X_train_sc, Y_train)
Y_pred = bnb.predict(X_test_sc)
print(classification_report(Y_test, Y_pred))
```

```python
# SVM --> SVC (Support Vector Classifier)
from sklearn.svm import SVC
svc = SVC(class_weight='balanced', random_state=7)
svc.fit(X_train_sc, Y_train)
Y_pred = svc.predict(X_test_sc)
print(classification_report(Y_test, Y_pred))
```

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_sc, Y_train)
Y_pred = knn.predict(X_test_sc)
print(classification_report(Y_test, Y_pred))
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion='gini',class_weight='balanced',random_state=7)
dtc.fit(X_train_sc, Y_train)
Y_pred = dtc.predict(X_test_sc)
print(classification_report(Y_test, Y_pred))
```

```python
def evaluate_model(model, X_train, X_test, Y_train, Y_test):
  model.fit(X_train, Y_train)
  Y_pred = model.predict(X_test)
  print(f"Classification report of {model} : \n{classification_report(Y_test, Y_pred)}")
```

```python
models = [LogisticRegression(class_weight='balanced', random_state=7),
          GaussianNB(),
          BernoulliNB(),
          SVC(class_weight='balanced', random_state=7),
          DecisionTreeClassifier(criterion='gini',class_weight='balanced',random_state=7),
          KNeighborsClassifier(n_neighbors=5)]
for i in models:
  evaluate_model(i, X_train_sc, X_test_sc, Y_train, Y_test)
```
