```python
from pandas import read_csv
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input
```

```python
df = pd.read_csv("/content/BankChurners.csv")
df.shape
```

```python
df.info()
```

```python
df.drop(['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',
         'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2',
         'CLIENTNUM'],axis=1,inplace=True)
```

```python
df.shape
```

```python
df['Attrition_Flag'].unique()
```

```python
label_map = {'Attrition_Flag':{'Existing Customer':0,
                               'Attrited Customer':1}}
df.replace(label_map,inplace=True)
```

```python
df.info()
```

```python
df_ohe = pd.get_dummies(df)
```

```python
df_ohe.shape
```

```python
#Split X & Y
X = df_ohe.drop(['Attrition_Flag'], axis=1)
Y = df_ohe['Attrition_Flag']
```

```python
X.shape
```

```python
Y.shape
```

```python
Y.value_counts(normalize=True)
```

```python
# split into train and test datasets
X_train, X_test, y_train, y_test = train_test_split(X, Y,
                                                    test_size=0.3,
                                                    random_state=7,
                                                    stratify=Y)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
n_features = X_train.shape[1]
```

```python
# define model
model1 = Sequential()
model1.add(Input((n_features,)))
model1.add(Dense(5, activation='relu'))
model1.add(Dense(4, activation='tanh'))
model1.add(Dense(20, activation='linear'))
model1.add(Dense(14, activation='gelu'))
model1.add(Dense(25, activation='sigmoid'))
model1.add(Dense(1, activation='sigmoid')) #output layer
```

## In model Summary Output shape is None, which represent Batch Size


## For Every layer Weight & Bias
- Layer 1 :  37 input X 5 output --> W[5 x 37] B[5 x 1] : 185 + 5 --> `190`

- Layer 2 :  5 input X 4 output --> W[4 x 5] B[4 x 1] --> `24`

- Layer 3 :  4 input X 20 output --> W[20 x 4] B[20 x 1] --> `100`

- Layer 4 :  20 input X 14 output --> W[14 x 20] B[14 x 1] --> `294`

- Layer 5 :  37 input X 5 output --> W[5 x 37] B[5 x 1] --> `375`

- Layer 6 :  25 input X 1 output --> W[1 x 25] B[1 x 1] --> `26`

- Total Parameter :-> `1009`




```python
model1.summary()
```

## Model.compile -->  Sets the optimiser, learning rate & Loss function(Cost Function)

```python
# compile the model
model1.compile(optimizer='adam', loss='binary_crossentropy',
               metrics=['accuracy','precision','recall','f1_score'])
```

- Shuffle the data, so that model doesn't Overfit
- For Randomness

- fit --> weight initialization (By Default Xavier Uniform) --> Forward Propagation --> Loss Calculation --> Back Propagate (Adam Optimiser) runs tills the last batch (shuffling happens in between Once per Epoch)

- Validation Error --> After every Epoch


```python
# fit the model (traning)
history = model1.fit( X_train, y_train, epochs=10,
                     batch_size=32,
                     validation_split=0.3,
                      verbose=2,shuffle=True)
```

```python
df.drop(['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',
         'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2',
         'CLIENTNUM'],axis=1,inplace=True)
```

```python
df_ohe = pd.get_dummies(df)
```

```python
df_ohe.shape
```

```python
df_ohe.info()
```

```python
X = df_ohe.drop(['Attrition_Flag_Attrited Customer',
                 'Attrition_Flag_Existing Customer'], axis=1)
Y = df_ohe[['Attrition_Flag_Attrited Customer',
                 'Attrition_Flag_Existing Customer']]
```

```python
X.shape,Y.shape
```

```python
# split into train and test datasets
X_train, X_test, y_train, y_test = train_test_split(X, Y,
                                                    test_size=0.3,
                                                    random_state=7,
                                                    stratify=Y)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
n_features = X_train.shape[1]
```

```python
# define model
model1 = Sequential()
model1.add(Input((n_features,)))
model1.add(Dense(5, activation='relu'))
model1.add(Dense(2, activation='softmax'))
```

```python
model1.compile(optimizer='adam', loss='binary_crossentropy',
               metrics=['accuracy','precision','recall','f1_score'])
```

```python
# fit the model (traning)
history = model1.fit( X_train, y_train, epochs=5, batch_size=32,
                     validation_split=0.3, verbose=2,
                      shuffle=True)
```
