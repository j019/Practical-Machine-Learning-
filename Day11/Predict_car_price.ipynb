```python
import pandas as pd
```

```python
df = pd.read_csv("/content/car_price_prediction_regression.csv")
```

```python
df.shape
```

```python
df.info()
```

```python
df_ohe = pd.get_dummies(df)
```

```python
df_ohe.shape
```

```python
df_ohe['Price'].skew()
```

# X & Y split

```python
X= df_ohe.drop('Price', axis=1)
y= df_ohe['Price']
```

```python
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,
                                                 random_state=7)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

# Create NN model for regression

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
n_features = X_train.shape[1]
n_features
```

```python
# define model
model1 = Sequential()
model1.add(Dense(5, activation='relu', input_shape=(n_features,)))
model1.add(Dense(100, activation='relu'))
# model1.add(Dense(800, activation='relu'))
# model1.add(Dense(8, activation='relu'))
# model1.add(Dropout(0.2))
model1.add(Dense(1, activation='linear'))
```

```python
model1.summary()
```

```python
model1.compile(optimizer='adam',loss="mse", metrics=['mae'])
```

```python
# fit the model (training)
history = model1.fit( X_train, y_train, epochs=100, batch_size=32,
                     validation_split=0.3, verbose=2,shuffle=False)
```

```python
loss, mae = model1.evaluate(X_test, y_test, verbose=0)
print('Test mae: %.3f' % mae)
```
